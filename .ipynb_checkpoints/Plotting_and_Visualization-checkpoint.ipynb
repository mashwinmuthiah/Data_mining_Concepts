```python
from numpy.random import randn
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc = 'best')
```

```python
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv(r'C:\Users\ashwi\Desktop\Ashwin\pydata-book\examples/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style='k-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
lab = ['1','6','3','5','8','1','3','9']
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot([1,6,3,5,8,1,3,9])

for i,j in enumerate([1,6,3,5,8,1,3,9]):
    ax.annotate(lab[i],xy = (i,j),xytext = (i,j+1),arrowprops = dict(facecolor='red',edgecolor = 'blue', headwidth=4, width=2,
                                headlength=4))
plt.savefig(r'C:\Users\ashwi\Desktop\Ashwin\data_mining_concepts\check.pdf')
```

```python
tips = pd.read_csv(r'C:\Users\ashwi\Desktop\Ashwin\pydata-book\examples/tips.csv')
tips
```

```python
cross_tips = pd.crosstab(tips['day'], tips['size'])
```

```python
cross_tips= cross_tips.loc[:,2:6]
cross_tips
```

```python
tips_a=cross_tips.div(cross_tips.sum(1), axis=0)
fig,axes = plt.subplots(1,1)
tips_a.plot.bar(ax = axes)
axes.set_title("Plot")
```

```python
 macro = pd.read_csv(r'C:\Users\ashwi\Desktop\Ashwin\pydata-book\examples/macrodata.csv')

data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]

trans_data = np.log(data).diff().dropna()

trans_data[-5:]
```

```python
sns.regplot(x = 'cpi',y = 'm1',data=trans_data)
```

```python
sns.pairplot(trans_data,diag_kind='kde', plot_kws={'alpha': 0.2},diag_kws=dict(shade=True))
```

```python
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker',
   .....:                kind='bar', data=tips[tips.tip_pct < 1])
sns.factorplot(x='day', y='tip_pct', row='time',
   .....:                col='smoker',
   .....:                kind='bar', data=tips[tips.tip_pct < 1])
```
